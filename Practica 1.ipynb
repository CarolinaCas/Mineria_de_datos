### ?df.dropna() 
DataFrame.dropna(self, axis=0, how='any', thresh=None, subset=None, inplace=False)
Esta función elimina los valores faltantes, regresa un DataFrame sin las entradas de NA.

Parametros:
- axis{0 or ‘index’, 1 or ‘columns’}, default 0 
    Determina si se van a remover las filas o columnas.
    0, or ‘index’ : da las filas que contienen valores faltantes.
    1, or ‘columns’ : da las columnas que contienen valores faltantes.
- how{‘any’, ‘all’}, default ‘any’
    Determina si la fila o la columna es removida del DataFrame, cuando tiene por lo menos un NA o todos NA.
    ‘any’ : si hay un valor NA presente, da esa fila o columna.
    ‘all’ :si todos los valores son NA, da esa fila o columna.
- threshint, optional
    Requiere cuantos no NA valores.
- subsetarray-like, optional
    Las etiquetas en el eje a considerar.
- inplacebool, default False
    Si es verdadero, entonces hace la operacion en el lugar y regresa None.

DataFrame.describe(self: ~FrameOrSeries, percentiles=None, include=None, exclude=None)

Esta función genera estadisticas descriptivas del DataFrame. Incluyen esas que resumen la tendencia central, la dispersion y la forma de la distribucion del set de datos expluyendo los valores NA. Analiza ambos, los valores numericos y los objetos, asi como los sets de columnas de los DataFrames con datos mezclados. La salida va a variar dependiendo en que datos se le den de entrada. 

Parametros:
- percentileslist-like of numbers, optional
    Los percentiles se incluyen opcionalmente en la salida. Todos deben de caer entre 0 y 1. El predeterminado es  [.25, .5, .75], que regresa el 25°, 50° y 75° percentiles.

- include‘all’, list-like of dtypes or None (default), optional
    Una lista de los tipos de datos se inluye en el resultado. Es ignorado para las series. Hay varias opciones:
        ‘all’ : Todas las columnas de entrada seran incluidas en la salida de información.
        A list-like of dtypes : El limite de los resultados de los datos provistos. Para limitar el resultado a los tipos numericos: numpy.number. Para limitar a objetos: numpy.object. Para seleccionar datos categoricos usar: 'category'.None (default) : El resultado va a incluir todas las columnas.

- excludelist-like of dtypes or None (default), optional,
    Una lista de datos a omitir en el resultado. Series a ignorar.
        A list-like of dtypes : exluye los tipos de datos del resultado. PAra exluir numericos: numpy.number, para excluir columnas: numpy.object, para datos categoricos use: 'category'. None (default) : No excluye nada.
